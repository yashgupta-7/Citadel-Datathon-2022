```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import gc
%matplotlib inline

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

DATA_DIR = '/content/drive/MyDrive/datathon_2022/data'
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df_rej = pd.read_pickle(os.path.join(DATA_DIR, 'Lending_Club_Rejected_2014_2018.pickle'))

df_rej['Application_Date'] = pd.to_datetime(df_rej['Application_Date'])
df_rej['year'] = pd.DatetimeIndex(df_rej['Application_Date']).year.astype(int)
df_rej['Debt_to_Income_Ratio'] = df_rej['Debt_to_Income_Ratio'].apply(lambda x: float(x.strip('%')))
df_rej.loc[df_rej['Debt_to_Income_Ratio'] < 0, 'Debt_to_Income_Ratio'] = np.nan

df_rej['zip3'] = df_rej['Zip_Code'].apply(lambda x: str(x)[:3])
df_rej = df_rej.drop(columns=['Loan_Title', 'Risk_Score', 'Zip_Code', 'Policy_Code'])
```

```python
df_rej.rename(columns = {'Amount_Requested': 'loan_amnt', 'Application_Date': 'issue_d', 
                         'Debt_to_Income_Ratio': 'dti', 'State': 'addr_state', 
                         'Employment_Length': 'emp_length'}, inplace=True)
```
