```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/FRB_G19.csv')
```

```python
df_1 = df.loc[1, :]
```

```python
df_2 = df.loc[7:, :]
```

```python
df_2['date'] = pd.to_datetime(df_2['Series Description'])
```

```python
mask = (df_2['date'] >= '2014-1') & (df_2['date'] <= '2018-12')
df_3 = df_2[mask]
df_3 = df_3.set_index('date')
# del df_3['Series Description']
col = 'Total consumer credit owned by depository institutions, seasonally adjusted level'
df_3[col] = df_3[col].astype(float)
df_3.plot(y =col, figsize=(10, 10))
```

```python
for c in df_3.columns:
    print(c)
```
